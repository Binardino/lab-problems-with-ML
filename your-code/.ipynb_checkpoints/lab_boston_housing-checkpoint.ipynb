## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    R2D2 = r2_score(y_true, y_predict)
    
    return R2D2
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
from sklearn.model_selection import train_test_split
```

```python
medv_target = boston['medv']
```

```python
features_boston = boston.drop(columns='medv')
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_boston, medv_target, test_size=0.2, shuffle=True)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
```

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
list_depth = [2, 4, 6, 8, 10]
```

```python
# Creating Five separate RFR Models here with the given max depths
#training both on X_Train & X_Test
```

```python
# Creating a for loop for each elem in list_depth to test the RandomForest on each max_depth parameter (for train & test sets)
# then appending the results to 2 different lists to then compare & plot their differences in performance score

rf_X_train_score = []
rf_X_test_score = []

for i in list_depth:
    rf_train = RandomForestRegressor(max_depth = i , random_state=42)
    rf_train.fit(X_train, y_train)
    ytrain_predict = rf_train.predict(X_train)
    print(f'the accuracy of the RF Train with a max depth of {rf_train.max_depth} is of',performance_metric(y_train, ytrain_predict))
    rf_X_train_score.append(performance_metric(y_train, ytrain_predict))

    rf_test = RandomForestRegressor(max_depth = i , random_state=42)
    rf_test.fit(X_test, y_test)
    ytest_predict = rf_train.predict(X_test)
    print(f'the accuracy of the RF Test with a max depth of {rf_test.max_depth} is of',performance_metric(y_test, ytest_predict))
    rf_X_test_score.append(performance_metric(y_test, ytest_predict))
    print('-------------')
```

```python
rf_X_train_score
```

Now, plot the score for each tree on the training set and on the testing set.

```python
plt.figure(figsize=(20,10))
plt.plot(list_depth, rf_X_train_score, c='g', label='perf score train model')
plt.plot(list_depth, rf_X_test_score, c='r' ,label='perf score test model')
```

What do these results tell you about the effect of the depth of the trees on the performance of the model?

```python
# Up to 4 in max_depth, there is an important difference between the two models ; after 4 levels of depth, the test set RF
#reached a plateau in performance, which indicates that adding more depth does not increase the accuracy

# the train set RF keeps increasing its accuracy past 4 levels of depth, which seems to indicate overfitting for the train set RF
```

### Bias-Variance Tradeoff
When the model is trained with a maximum depth of 1, does the model suffer from high bias or from high variance? How about when the model is trained with a maximum depth of 10?

```python
rf1 = RandomForestRegressor(max_depth=1)
rf1.fit(X_train, y_train)
```

```python
y_predict1 = rf1.predict(X_test)
print (f'the accuracy of the random forest for  {rf1.max_depth} is of',performance_metric(y_test, y_predict1))
```

```python
rf10 = RandomForestRegressor(max_depth=10)
rf10.fit(X_train, y_train)
```

```python
y_predict10 = rf10.predict(X_test)
```

```python
print (f'the accuracy of the random forest for {rf10.max_depth} is of',performance_metric(y_test, y_predict10))
```

#### The performance metrics is much higher when the Random forest is tested with 10 max_depth

### Best-Guess Optimal Model
What is the max_depth parameter that you think would optimize the model? Run your model and explain its performance.

```python
rf_i = []
for i in range(1,13):
    rf = RandomForestRegressor(max_depth=i)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    perf = performance_metric(y_test, y_pred)
    print(f'the performance with of the RFR with a max_depth of {rf.max_depth} is of', perf)
    rf_i.append(perf)
```

```python
plt.figure(figsize=(20,10))
plt.plot((range(1,13)), rf_i, c='g')
```
